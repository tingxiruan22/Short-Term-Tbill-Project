```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("3Mtbill_2017.csv")
df
```

```python
plt.plot(df['DATE'], df['int'])
plt.show()
```

```python
df['DATE'] = pd.to_datetime(df['DATE'])

# Replace dots with NaN and convert to numeric
df['DGS3MO'].replace('.', None, inplace=True)
df['DGS3MO'] = pd.to_numeric(df['DGS3MO'], errors='coerce')

# Forward fill NaN values
df['DGS3MO'] = df['DGS3MO'].ffill()

df.rename(columns={'DGS3MO': 'int'}, inplace=True)
```

```python
df['pct_change'] = df['int'].pct_change() 
df
```

```python
plt.plot(df['DATE'], df['pct_change'], linewidth=0.5)
plt.show()
```
